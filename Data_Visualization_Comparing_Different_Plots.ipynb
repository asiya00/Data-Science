```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
sns.distplot(df['horsepower'],kde=False,rug=True)
plt.show()
```

```python
sns.jointplot(df['wheel-base'],df['length'])
```

```python
sns.jointplot(df['wheel-base'],df['length'],kind = 'reg')
```

```python
#kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'], but res was passed.
sns.jointplot(df['wheel-base'],df['length'],kind='reg')
```

```python
sns.jointplot(df['wheel-base'],df['length'],kind='kde')
```

```python
sns.pairplot(df[['length','horsepower','average-mileage']],size=5)
```

```python
sns.stripplot(df['engine-type'],df['horsepower'])
```

```python
sns.stripplot(df['engine-type'],df['horsepower'], jitter=True)
```

```python
sns.swarmplot(df['engine-type'],df['horsepower'])
```

```python
sns.stripplot(df['body-style'],df['horsepower'])
```

```python
sns.boxplot(df['body-style'],df['horsepower'])
```

```python
sns.boxplot(df['body-style'],df['horsepower'],hue=df['num-of-cylinders'])
```

```python
sns.barplot(df['body-style'],df['horsepower'])
```

```python
sns.barplot(df['body-style'],df['horsepower'],hue=df['engine-type'])
```

```python
sns.countplot(df['body-style'],hue=df['engine-type'])
```

```python
sns.pointplot(df['engine-type'],df['horsepower'])
```

```python
#kind = point,box,violin,count,strip
sns.factorplot(x="engine-type",y="horsepower",hue="company",col="num-of-cylinders", data = df, kind= 'swarm')
```

```python
sns.lmplot(x='horsepower',y='price',data=df)
```
