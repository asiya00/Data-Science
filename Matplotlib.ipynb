```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Line** **Plot**

```python
x = np.arange(1,10)
y = 2 * x
x,y
```

```python
plt.plot(x,y)
```

```python
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.plot(x,y)
plt.show()
```

## Changing Line Aesthetics

```python
x = np.arange(1,10)
y = 2 * x
plt.plot(x,y,color='g',linestyle=':',linewidth='2')
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.show()
```

## Adding Two lines in the same plot

```python
x = np.arange(1,10)
y1 = 2 * x
y2 = 3 * x
```

```python
# linestyle '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x,y1,label = "line 1",color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,label = "line 2",color='r',linestyle='-',linewidth=3)
y3 = x * 4
plt.plot(x,y3,label = "line 3",color='y',linestyle='--',linewidth=3)
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.grid(True)
plt.legend()
plt.show()
```

## Adding Subplots

```python
x = np.arange(1,11)
y1 = 2 * x
y2 = 3 * x
y3 = 4 * x
y4 = 5 * x
x,y1,y2,y3,y4
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(2,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.subplot(2,2,3)
plt.plot(x,y3,color='y',linestyle='-',linewidth=3)
plt.subplot(2,2,4)
plt.plot(x,y4,color='blue',linestyle='-',linewidth=3)
```

## **Bar** **Plot**

```python
Students = {'Bob':87, "Matt":56, "Sam":27}
names = Students.keys()
values = Students.values()
plt.bar(names,values)
plt.title('Bar Plot')
plt.xlabel('Names')
plt.ylabel('Marks') 
plt.grid(True)
plt.show()
```

## Horizontal Bar plot

```python
Students = {'Bob':87, "Matt":56, "Sam":27}
names = list(Students.keys())
values = list(Students.values())
plt.barh(names,values)
```

## **Scatter** **Plot**

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='*',c='g',s=100)
```

## Adding two markers in the same plot

```python
x = [10,20,30,40,50,60,70,80,90]
y1 = [8,1,7,2,0,3,7,3,2]
y2 = [7,2,5,6,9,1,4,5,3]
y3 = [7,2,5,6,9,1,4,5,3]
plt.scatter(x,y1,label='Y1',marker='*',c='y',s=100)
plt.scatter(x,y2,label='Y2',marker='^',c='g',s=200)
plt.scatter(x,y3,label='Y3',marker='1',c='brown',s=200)
plt.legend()
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y1 = [8,1,7,2,0,3,7,3,2]
y2 = [7,2,5,6,9,1,4,5,3]
plt.subplot(1,2,1)
plt.scatter(x,y1,label='Y1',marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,y2,label='Y2',marker='.',c='r',s=200)
```

## Histogram

```python
data = [1,3,3,3,3,9,9,4,5,5,8,8,6,7]
plt.hist(data,color='g',bins=4)
plt.show()
```

## **Box** **Plot**

```python
x = [1,2,3,4,5,6,7,8,9]
y1 = [1,2,3,4,5,4,3,2,1]
y2 = [6,7,8,9,8,7,6,5,4]
data = list([x,y1,y2])
plt.boxplot(data)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y1 = [1,2,3,4,5,4,3,2,1]
y2 = [6,7,8,9,8,7,6,5,4]
data = list([x,y1,y2])
plt.violinplot(data,showmedians = True,vert=True,showmeans=True)
plt.show()
```

## **Pie**-**Chart**

```python
Fruits = ['Apple','Mango','Banana','Orange']
Quantity = [67,34,100,29]
plt.pie(Quantity,labels=Fruits,colors=['Red','Yellow','Green','Orange'])
```

```python
plt.pie(Quantity,labels=Fruits,autopct = '%0.1f%%')
```

## **DoughNut**-**Chart**

```python
Fruits = ['Apple','Mango','Banana','Orange']
Quantity = [67,34,100,29]
plt.pie(Quantity,labels=Fruits,radius=2,autopct = '%0.1f%%')
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
